```
# DPLL

from re import sub
from math import fabs
from copy import deepcopy
import timeit

valor_verdade = []
unit_propa = 0
splitting = -1

def remove_literais(fnc):
    aux_fnc = []
    literais = set()

    for i in range(len(fnc)):
        aux_fnc.append(fnc[i].split())

    for i in range(len(aux_fnc)):
        for j in range(len(aux_fnc[i])):
            literais.add(str(int(fabs(int(aux_fnc[i][j])))))

    return list(literais)

def separa_literais(arquivo_fnc):
    arquivo_fnc = arquivo_fnc.splitlines()
    fnc = []
    literais = set()

    for i in range(len(arquivo_fnc)):
        fnc.append(arquivo_fnc[i].split())

    for i in range(len(fnc)):
        for j in range(len(fnc[i])):
            fnc[i][j] = int(fnc[i][j])
            literais.add(str(int(fabs(fnc[i][j]))))

    literais = list(literais)
    return arquivo_fnc, literais

def dpll(fnc, literais):
    global valor_verdade, unit_propa, splitting

    splitting += 1

    # ------------start unit propa-----------------
    while True:
        if len(literais) == 1 and len(fnc) > 1:
            # print('--------------------------------------------------')
            # print(fnc)
            # print(literais)
            break

        delete_list = []
        clausula_unitaria = None
        fnc = list(set(fnc))
        # print('--------------------------------------------------')
        #print('FNC : ', fnc)
        #print('Literais : ', literais)

        # Encontra cláusula unitária.
        for i in range(len(fnc)):
            if ' ' not in fnc[i]:
                clausula_unitaria = fnc[i]
                valor_verdade.append(clausula_unitaria)
                #print(f'Unit : {clausula_unitaria}')
                del fnc[i]
                break

        # Verifica se existe clausula_unitaria negada.
        if clausula_unitaria:
            for i in range(len(fnc)):
                if '-' in clausula_unitaria:
                    if clausula_unitaria.replace('-', '') == fnc[i]:
                        return False
                else:
                    if '-' + clausula_unitaria == fnc[i]:
                        return False

        if clausula_unitaria and '-' not in clausula_unitaria:
            unit_propa += 1
            for j in range(len(fnc)):
                if '-' + clausula_unitaria in fnc[j]:
                    if '-' + clausula_unitaria == fnc[j]:
                        delete_list.append(j)
                    else:
                        fnc[j] = fnc[j].replace('-' + clausula_unitaria, '')
                        fnc[j] = sub("\s+", " ", fnc[j].strip())

                elif clausula_unitaria in fnc[j]:
                    delete_list.append(j)

            for index in sorted(delete_list, reverse=True):
                del fnc[index]
            literais = remove_literais(fnc)

        elif clausula_unitaria and '-' in clausula_unitaria:
            unit_propa += 1
            for j in range(len(fnc)):
                if clausula_unitaria in fnc[j]:
                    delete_list.append(j)
                elif clausula_unitaria.replace('-', '') in fnc[j]:
                    if clausula_unitaria.replace('-', '') == fnc[j]:
                        delete_list.append(j)
                    else:
                        fnc[j] = fnc[j].replace(clausula_unitaria.replace('-', ''), '')
                        fnc[j] = sub("\s+", " ", fnc[j].strip())

            for index in sorted(delete_list, reverse=True):
                del fnc[index]
            literais = remove_literais(fnc)

        else:
            break
    # --------------end unit propa----------------------

    # checking clause
    if fnc == []:
        return True
    elif len(literais) == 1 and len(fnc) > 1:
        return False

    # case-splitting
    if dpll(fnc + [literais[0]], deepcopy(literais)) or dpll(fnc + ['-' + literais[0]], deepcopy(literais)):
        return True
    else:
        return False


def main():
    global valor_verdade
    input_cnf = open('clausulas.txt').read()
    fnc, literais = separa_literais(input_cnf)

    if dpll(fnc, literais):
        print('\nSatisfazivel!!!\n')

        # Imprimir Valoração verdadeira
        for i in range(len(valor_verdade) - 1, 0, -1):
            for j in range(i - 1, -1, -1):
                if '-' in valor_verdade[i]:
                    if valor_verdade[i] == valor_verdade[j] or valor_verdade[i].replace('-', '') == \
                            valor_verdade[j]:
                        valor_verdade[j] = ''
                else:
                    if valor_verdade[i] == valor_verdade[j]:
                        valor_verdade[j] = ''

        if '' in valor_verdade:
            valor_verdade = list(set(valor_verdade))
            valor_verdade.remove('')

        print('Valor verdade: ', end='')
        arquivo = open('solucao_sudoku.txt', 'w')
        for i in valor_verdade:
            if '-' not in i:
                print(f'|{i}', end='')
                arquivo.write(f'{i}\n')
        print('|')
        arquivo.close()

        print('\nPrint Sudoku resolvido:', end='\n\n')
        arquivo = open('solucao_sudoku.txt').read().splitlines()
        matriz = []
        for i in range(9):
            matriz = matriz + [[0]*9]

        for i in range(81):
            matriz[int(arquivo[i][0]) - 1][int(arquivo[i][1]) - 1] = arquivo[i][2]
        
        for i in range(9):
            for j in range(9):
                print(f'{matriz[i][j]} ', end='')
            print()   

    else:
        print('Insatisfazivel!!!')

    #print('[Unit propagation Count] : ', unit_propa)
    #print('[Splitting Count] : ', splitting)

if __name__ == '__main__':
    start = timeit.default_timer()

    main()

    stop = timeit.default_timer()
    print('\nTempo de execucao: ', stop-start)
```
